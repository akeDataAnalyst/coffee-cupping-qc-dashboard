#### Setup & Import Modules

```python
import pandas as pd
import numpy as np
import plotly.express as px
import sys
from pathlib import Path

# Add project root to path
PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

# Import our modules
from src.config import BASE_DIR
from src.data_loader import load_cupping_data, validate_cupping_data
from src.utils import get_score_summary, get_region_performance

print("Setup Complete")
print(f"Project Root: {BASE_DIR}")
```

#### Load and Validate Data

```python
df = load_cupping_data()
df = validate_cupping_data(df)

print(f"Loaded {len(df)} cupping records")
print(f"Date Range: {df['cupping_date'].min().date()} to {df['cupping_date'].max().date()}")
print(f"Average Total Score: {df['total_score'].mean():.2f}")
```

#### Overall Score Distribution

```python
print("=== OVERALL CUPPING SCORE DISTRIBUTION ===\n")

score_stats = {
    'Total Samples': len(df),
    'Average Score': df['total_score'].mean().round(2),
    'Highest Score': df['total_score'].max(),
    'Lowest Score': df['total_score'].min(),
    'Outstanding (88+)': (df['total_score'] >= 88).sum(),
    'Excellent (85+)': (df['total_score'] >= 85).sum()
}

for key, value in score_stats.items():
    print(f"{key:25}: {value}")

# Visualization
fig = px.histogram(df, x='total_score', nbins=15, color='quality_tier',
                  title="Distribution of SCA Cupping Scores",
                  labels={'total_score': 'Total SCA Score'})
fig.show()
```

#### Quality Tier Analysis

```python
print("=== QUALITY TIER ANALYSIS ===\n")

tier_summary = get_score_summary(df)
display(tier_summary)

fig = px.pie(df, names='quality_tier', title="Proportion by Quality Tier")
fig.show()
```

#### Regional Performance

```python
print("=== REGIONAL PERFORMANCE ===\n")

region_perf = get_region_performance(df)
display(region_perf)

fig = px.bar(
    df.groupby('region')['total_score'].mean().reset_index(),
    x='region', y='total_score',
    title="Average Cupping Score by Region",
    color='region'
)
fig.show()
```

#### Processing Method Analysis

```python
print("=== PROCESSING METHOD ANALYSIS ===\n")

processing_analysis = df.groupby('processing_method').agg({
    'total_score': ['mean', 'max', 'count'],
    'defects_per_300g': 'mean',
    'aroma': 'mean',
    'acidity': 'mean',
    'body': 'mean'
}).round(2)

display(processing_analysis)
```

#### Attribute Correlation & Balance

```python
print("=== ATTRIBUTE ANALYSIS ===\n")

# Correlation between attributes
attributes = ['aroma', 'flavor', 'aftertaste', 'acidity', 'body', 'balance']
corr = df[attributes].corr().round(2)

print("Attribute Correlation Matrix:")
display(corr)

# Top 5 Highest Scoring Samples
print("\n=== TOP 5 HIGHEST SCORING CUPS ===")
top5 = df.nlargest(5, 'total_score')[[
    'sample_id', 'lot_id', 'region', 'processing_method', 
    'total_score', 'flavor_notes'
]]
display(top5)
```

#### Key Insights & Recommendations

```python
print("=== KEY INSIGHTS FOR VOLCAFE SPECIALTY TEAM ===\n")

print(f"1. Overall Average Score          : {df['total_score'].mean():.2f}")
print(f"2. Outstanding Samples (88+)     : {(df['total_score'] >= 88).sum()}")
print(f"3. Best Performing Region        : {df.groupby('region')['total_score'].mean().idxmax()}")
print(f"4. Best Processing Method        : {df.groupby('processing_method')['total_score'].mean().idxmax()}")
print(f"5. Average Defects               : {df['defects_per_300g'].mean():.1f} per 300g")

print("\n--- Recommendations ---")
print("• Prioritize Yirgacheffe and Sidama washed lots scoring 88+")
print("• Focus on suppliers delivering consistent clean cup and sweetness")
print("• Natural processed coffees also show strong potential (especially in Sidama)")
print("• Continue monitoring defects closely to maintain ECX Grade 1 standards")
```
